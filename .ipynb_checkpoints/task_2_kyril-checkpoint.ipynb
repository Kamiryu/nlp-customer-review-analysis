#### Normalize text

```python
# Define stopwords and punctuation
stop_words = set(stopwords.words('english'))
punctuation = set(string.punctuation)

def clean_tokens(text : str, lem : bool = False, pos_arg={"VERB":'v', "ADJ":'a', "ADV":'r', "NOUN":'n'}) -> List:
    
    # Tokenize the text
    token = word_tokenize(text.lower())
    # Remove unwanted punctuation and stop words
    # if extra_clean == True:
    #    tokens = [word for word in tokens if (word not in punctuation and word not in stop_words)]

    if lem == True:
        
        output = []
        
        for w, pos in nltk.pos_tag(token, tagset="universal"):
            pos = pos_arg[pos] if pos in pos_arg.keys() else 'n'
            lem = wnl.lemmatize(w, pos=pos)

            if lem not in stopwords:
                output.append(lem)
    
        return output

    return token
```

#### Extract only food related stuff

```python
def is_food_related(word : str) -> bool:

    lemmatizer = WordNetLemmatizer()
    lemword = lemmatizer.lemmatize(word.lower())
    
    restaurant_keywords = ["restaurant", "dining", "menu", "waiter", "chef", "kitchen", "cafe", "bistro", "buffet", "canteen", "diner", "american"]
    cuisine_types = ["italian", "mexican", "chinese", "japanese", "indian", "french", "thai", "greek", "spanish", "mediterranean", "vietnamese", "korean"]

    lem_restaurant_keywords = [lemmatizer.lemmatize(keyword) for keyword in restaurant_keywords]
    lem_cuisine_types = [lemmatizer.lemmatize(keyword) for keyword in cuisine_types]
    
    # Check if the word is in the predefined list
    if any(lemword in lr for lr in lem_restaurant_keywords + lem_cuisine_types):
        return True

    # Extract from nltk
    synsets = wordnet.synsets(word, pos=wordnet.NOUN)
    for synset in synsets:
        if 'food' in synset.lexname():
            return True 

        if any(hypernym in synset.hypernyms() for hypernym in [wordnet.synset('restaurant.n.01'), wordnet.synset('eating_place.n.01')]):
            return True 

    return False

```
